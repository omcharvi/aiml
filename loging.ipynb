```python
import logging
logging.basicConfig(filename='prog.txt', level=logging.DEBUG, format='%(asctime)s - %(levelname)s - %(message)s')
try :
    logging.info("Program started")
    x = int(input("Enter a number: "))
    logging.info(f"User entered: {x}")
    y = int(input("Enter another number: "))
    logging.info(f"User entered: {y}")
    result = x / y
    logging.info(f"Result of division: {result}")
except Exception as e:
    logging.error(f"An error occurred : {e}")
else :
    logging.info("Program ended successfully")
finally:
    logging.debug("Execution completed")
```
