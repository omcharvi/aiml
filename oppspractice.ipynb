```python
class Employee:
    def __init__(self, emp_id, salary):
        self.emp_id = emp_id
        self.salary = salary

    def __str__(self):
        return f" {self.emp_id}-{self.salary}"
    

e1 = Employee("Dularchand Mahto", 45000)
e2 = Employee("Kashiram Mahto", 55000)
e3 = Employee("Ramnaraya Gandhi", 60000)

all_empl = [e1,e2,e3]
tot_sal = sum(empl.salary for empl in all_empl)

print("Total salary details")
for i in all_empl:
    print(i)

print(f"Total salary to be paid : ₹ {tot_sal}")
```

```python
class Flight:
    def __init__(self,flight_no, source, destination):
        self.flight_no = flight_no 
        self.source = source
        self.destination = destination

    def __str__(self):
        return f"{self.flight_no}-{self.source}- {self.destination}"
    
f1 = Flight("6E546", "Patna" , "Gaya")
f2 = Flight("6E6860", "Jamshedpur", "Delhi")
f3 = Flight ("8983", "Delhi","Gaya")
f4 = Flight ("6E567", "Delhi", "Bangalore")

all_flight = [f1,f2,f3,f4]

print("Flight Details from Delhi")

for flight in all_flight:
    if flight.source == "Delhi":
        print(flight)
```

```python
class Weather :
    def __init__(self, city, temperature, condition):
        self.city = city
        self.temperature = temperature
        self.condition = condition

    def __str__(self):
        return f" {self.city}- {self.temperature} .C - {self.condition}"
    

c1 = Weather("Bangalore", 26, "Cloudy to rainy" )
c2 = Weather("Delhi", 30, "Mostly Humidity")
c3 = Weather("Chennai", 34, "Humidity and Heat")

all_city = [c1,c2,c3]
high_tem = max(all_city, key = lambda x :x.temperature)

print("Highest temperature in city is : ")
print(high_tem)
      
```
