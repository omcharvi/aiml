```python
class Product :
    def __init__(self, name, price):
        self.name = name
        self.__price = price

    @property
    def price (self):
        return self.__price

    @price.setter
    def price(self, value):
        if value < 0 :
            print("Price can not be negative")
        else:
            self.__price = value
            print(f"Price updated to {self.__price}")


    def show (self):
        print(f"Product :{self.name},Price: {self.__price}")


p1 = Product("Samosha", 20)
print("Current price:", p1.price)
p1.price = 25
p1.price = -10
p1.show()

```

```python
def timer(func):
    def wrapper():
        start_time = time.time()
        func()
        end_time = time.time()
        print(f"function executed in {end_time -start_time} seconds")
    return wrapper
```

```python
import time
def time_it(func):
    def wrapper():
        start_time = time.time()
        func()
        end_time = time.time()
        print(f"Execution time :{start_time - end_time} seconds")
    return wrapper
```

```python
@time_it
def slow_function():
    for i in range(5):
        print(f"iteration {i+1} started")
        time.sleep(1)
    print(f"Function executed!")
```

```python
def require_login(func):
    def wrapper(*args, **kwargs):
        logged_in = kwargs.get('loged_in', False)

        if not logged_in :
            print(f"Access denied ! please login first")
            return None
        else:
            print("ACcess granted")
            return func(*args, **kwargs)
    return wrapper 

```

```python
@require_login
def view_dasboard(*args, **kwargs):
    print("Welcomeo to your dashboard!")

@require_login
def edit_profile(*args, **kwargs):
    print("Profile in editing progress.....kindly wait:)")
```
